```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('student-por.csv')
print(df)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
print("\nColumn data types:")
print(df.dtypes)
```

```python
df.shape
```

```python
average_math_score = df['G3'].mean()
print("\nAverage score in math (G3):", average_math_score)
```

```python
students_above_15 = df[df['G3'] > 15].shape[0]
print("Number of students scoring above 15 in G3:", students_above_15)
```

```python
correlation = df['studytime'].corr(df['G3'])
print("Correlation between study time and final grade (G3):", correlation)
```

```python
average_grade_by_gender = df.groupby('sex')['G3'].mean()
print("Average final grade by gender:")
print(average_grade_by_gender)
```

```python

# Histogram of final grades
plt.figure(figsize=(8, 5))
sns.histplot(df['G3'], kde=True, bins=10, color='skyblue')
plt.title('Distribution of Final Grades (G3)', fontsize=14)
plt.xlabel('Final Grade (G3)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Scatter plot of study time vs final grade
plt.figure(figsize=(8, 5))
sns.scatterplot(x='studytime', y='G3', data=df, color='green')
plt.title('Study Time vs Final Grade (G3)', fontsize=14)
plt.xlabel('Study Time', fontsize=12)
plt.ylabel('Final Grade (G3)', fontsize=12)
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

```python

# Bar chart of average scores by gender
plt.figure(figsize=(8, 5))
sns.barplot(x='sex', y='G3', hue='sex', data=average_grade_by_gender, palette='coolwarm')
plt.title('Average Final Grade by Gender', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Average Final Grade (G3)', fontsize=12)
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'], fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```
